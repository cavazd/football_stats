```python
import pandas as pd
from pathlib import Path
```

```python
cwd = Path.cwd()
week = "week1"
data = cwd / week / "res"

csvs = data.glob("*.csv")

dfs = []
for csv in csvs:
  dfs.append(pd.read_csv(csv))


if len(dfs) == 0:
  exit()
elif len(dfs) == 1:
  df = dfs[0]
else:
  df = dfs[0]
  for partial_df in dfs[1:]:
    df = pd.concat([df, partial_df])
```
